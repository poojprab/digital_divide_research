```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

numerical_columns = [
    'Average_Rating',
    'Median income (dollars)_Families',
    'Total!!Less than $10,000_Families',
    'Total!!$10,000 to $14,999_Families',
    'Total!!$15,000 to $24,999_Families',
    'Total!!$25,000 to $34,999_Families',
    'Total!!$35,000 to $49,999_Families',
    'Total!!$50,000 to $74,999_Families',
    'Total!!$75,000 to $99,999_Families',
    'Total!!$100,000 to $149,999_Families',
    'Total!!$150,000 to $199,999_Families',
    'Total!!$200,000 or more_Families',
    'Total:!!Has a computer:!!With a broadband subscription:'
]

# imputing the NaN data by replacing it with the mean of the rest of the numbers in the column
imputer = SimpleImputer(strategy='mean')
data[numerical_columns] = data[numerical_columns].apply(pd.to_numeric, errors='coerce')
data_imputed = pd.DataFrame(imputer.fit_transform(data[numerical_columns]), columns=numerical_columns)

# PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data_imputed)

# Add the components back to the DataFrame
data['PC1'] = principal_components[:, 0]
data['PC2'] = principal_components[:, 1]

# Visualize the PCA results
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', data=data)
plt.title('PCA - 2 Principal Components')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the distortion (inertia) for a range of k values
inertia = []
k_range = range(1, 11)  # Test k from 1 to 10
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_imputed)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_range)
plt.grid()
plt.show()
```

```python
# Choose the optimal number of clusters (for example, k=3)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data['Cluster'] = kmeans.fit_predict(data_imputed)  # Add cluster labels to the dataframe
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=data, palette='viridis', s=100)
plt.title('K-Means Clustering on PCA Results')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```

```python
# Analyze cluster characteristics
cluster_summary = data.groupby('Cluster').mean(numeric_only=True)  # Only include numeric columns
print(cluster_summary)
```
